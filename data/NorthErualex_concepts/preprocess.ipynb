```python
import pandas as pd
```

```python
data = pd.read_csv("semantic-domains-and-their-concepts.csv", index_col=0).reset_index(drop=True)
data
```

```python
import json

# Load the data
with open("en_ru_de.json") as f:
    en_ru_de = json.load(f)
```

```python
for concept in data["English Word Form"]:
    not_found = True
    for lang in en_ru_de['rows']:
        if concept.lower() == lang['concepticon_name'].lower():
            not_found = False
            break

    if not_found:
        print(concept)
```
